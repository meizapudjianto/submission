## Exploratory Data Analysis (EDA)

```python
bike_df = hour_df.merge(day_df, on='dteday', how='inner', suffixes=('_hour', '_day'))
bike_df.head()
```

```python
weather_labels = {
    1: 'Cerah',
    2: 'Berawan',
    3: 'Hujan Ringan',
    4: 'Hujan Deras'
}

bike_df['weather_label'] = bike_df['weathersit_day'].map(weather_labels)
```

```python
bike_df.describe(include="all")
```

```python
bike_df.groupby('weather_label')['cnt_day'].mean().reset_index().sort_values("cnt_day")
```

```python
bike_df.groupby('hr')['cnt_hour'].mean().reset_index()
```

```python
bike_df.groupby('holiday_day')['cnt_day'].mean().reset_index().sort_values("cnt_day")
```

```python
bike_df.registered_hour.sum()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

Berapa jumlah jam penyewaan sepeda pada setiap bulannya?

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.barplot(data=day_df, x='mnth', y='cnt', ax=ax, color='#1f77b4')
ax.set(title='Jumlah Jam Penyewaan Sepeda Setiap Bulan')
```

### Pertanyaan 2:

Berapa rata-rata penyewaan sepeda perjam?

```python
sewa_jam = bike_df.groupby('hr')['cnt_hour'].mean()

fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(sewa_jam.index, sewa_jam.values, color='#1f77b4')  # Menggunakan ax.bar untuk subplot
ax.set(title='Rata - Rata Penyewaan per Jam', xlabel='Jam', ylabel='Rata - Rata')

plt.show()
```

## Conclusion

- Conclution pertanyaan 1 :
Berdasarkan diagram visualisasi diatas, penyewaan sepeda tertinggi pada bulan Juni.

- Conclution pertanyaan 2 :
Rata-rata penyewaan sepeda paling banyak terjadi pada jam 17 dan 18. Sedangkan paling sedikit jam 4.
